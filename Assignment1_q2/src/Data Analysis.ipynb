```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data_clean/clean_raw.csv")
df
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['reading score'], bins=20, kde=True)
plt.title('Histogram of scores')
plt.xlabel('reading scores')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='race/ethnicity', data=df)
plt.title('Bar Chart of parental level of education')
plt.xlabel('parental level of education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(x='parental level of education', y='writing score', data=df)
plt.title('Line Chart of  Scores Over parental level of education')
plt.xlabel('parental level of education')
plt.ylabel('writing Scores')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Boxplot of math score by Gender')
plt.xlabel('Gender')
plt.ylabel('math score')
plt.show()
```

```python
variables = ['writing score', 'reading score', 'math score', 'gender']  
sns.pairplot(df[variables], hue='gender', plot_kws={'alpha': 0.5})
plt.suptitle('Pair Plot of Scores by Gender', y=1.02) 
plt.show()
```
